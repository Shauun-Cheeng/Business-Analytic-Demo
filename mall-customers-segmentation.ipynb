```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Load datasets
df = pd.read_csv("/kaggle/input/mall-customers-segmentation/Mall_Customers.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Data distribution

x = df['Age']
y = df['Spending Score (1-100)']

# Change date to datetime
x = pd.to_datetime(x)
plt.scatter(x ,y , color='blue')

# Set the title and label
plt.title('Spending Score', fontsize="18")
plt.xlabel('Age', fontsize="14")
plt.ylabel('Spending Score (1-100)', fontsize="14")
```

```python
# Data distribution

x = df['Age']
y = df['Annual Income (k$)']
plt.scatter(x ,y , color='blue')

# Set the title and label
plt.title('Annual Income', fontsize="18")
plt.xlabel('Age', fontsize="14")
plt.ylabel('Annual Income (k$)', fontsize="14")
```

```python
# Data distribution

x = df['Annual Income (k$)']
y = df['Spending Score (1-100)']
plt.scatter(x ,y , color='blue')

# Set the title and label
plt.title('Spending Score', fontsize="18")
plt.xlabel('Annual Income (k$)', fontsize="14")
plt.ylabel('Spending Score (1-100)', fontsize="14")
```

```python
# Data distribution

sns.scatterplot(x= df['Annual Income (k$)'] ,y= df['Spending Score (1-100)'])

# Set the title and label
plt.title('Spending Score', fontsize="18")
plt.xlabel('Annual Income (k$)', fontsize="14")
plt.ylabel('Spending Score (1-100)', fontsize="14")
```

```python
# Regression model

sns.regplot(x= df['Annual Income (k$)'] ,y= df['Spending Score (1-100)'])
plt.title('Regression Model', fontsize="18")
```
